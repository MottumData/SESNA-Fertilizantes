# Challenging Cases: Duplicidad de localidades en un mismo Municipio y Entidad

```python
import pandas as pd
```

```python
inegi_2019 = pd.read_csv('../../data/dataset_inegi_clean_2019.csv')
inegi_2020 = pd.read_csv('../../data/dataset_inegi_clean_2020.csv')
inegi_2021 = pd.read_csv('../../data/dataset_inegi_clean_2021.csv')
inegi_2022 = pd.read_csv('../../data/dataset_inegi_clean_2022.csv')
```

## Análisis 2019

### Filtrando las entidades de interés 

```python
# Lista de entidades que quieres incluir en el filtro
entidades_interes = ['Morelos', 'Chiapas', 'Durango', 'Guerrero', 'Nayarit', 'Oaxaca', 'Puebla', 'Tlaxcala', 'Zacatecas']

# Filtrar el DataFrame por las entidades de interés
inegi_2019 = inegi_2019[inegi_2019['Entidad_inegi'].isin(entidades_interes)]
```

```python
print("Numero de rows: ", inegi_2019.shape)
```

### Duplicados 

```python
# Ordenar el DataFrame por Entidad_inegi, Municipio_inegi y Localidad_inegi
df_sorted = inegi_2019.sort_values(by=['Entidad_inegi', 'Municipio_inegi', 'Localidad_inegi'])

# Identificar duplicados basados en Entidad_inegi, Municipio_inegi y Localidad_inegi
duplicados = df_sorted[df_sorted.duplicated(subset=['Entidad_inegi', 'Municipio_inegi', 'Localidad_inegi'], keep=False)]
print("Número de rows con base en Entidad, Municipio y Localidad: " , duplicados.shape)
```

```python
duplicados
```

### Conteo de duplicados agrupado por Entidad, Municipio y Localidad
#### Hay un total de 4301 localidades que aparecen 2 o más veces en el dataset INEGI con fecha de corte Diciembre 2019

```python
# Contar duplicados agrupados por Entidad_inegi, Municipio_inegi y Localidad_inegi
conteo_duplicados = duplicados.groupby(['Entidad_inegi', 'Municipio_inegi', 'Localidad_inegi']).size().reset_index(name='Numero_Duplicados')

# Ordenar por el número de duplicados en orden descendente
conteo_duplicados = conteo_duplicados.sort_values(by='Numero_Duplicados', ascending=False)

# Mostrar el conteo de duplicados ordenado por count
conteo_duplicados
```

### Mostrando el ejemplo con mayor número de apariciones de un mismo nombre de Localidad pertenecientes al mismo Municipio y Entidad

```python
# Suponiendo que ya tienes el DataFrame de duplicados correctamente definido como 'duplicados'

# Filtrar duplicados donde Entidad_inegi sea 'Chiapas' y Localidad_inegi sea 'El Recuerdo'
duplicados[(duplicados['Entidad_inegi'] == 'Chiapas') & 
           (duplicados['Municipio_inegi'] == 'Villa Corzo')&
           (duplicados['Localidad_inegi'] == 'El Recuerdo')]
```
